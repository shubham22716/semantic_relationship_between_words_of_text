```
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize 
import warnings 
  
warnings.filterwarnings(action = 'ignore') 
  
import gensim 
from gensim.models import Word2Vec 
nltk.download('punkt')
```

```
from google.colab import files
files.upload()
```

```
sample = open("alice.txt") 
s = sample.read() 
```

```
f = s.replace("\n", " ") 
  
data = [] 
  
  
```

```
for i in sent_tokenize(f): 
    temp = [] 
      
    # tokenize the sentence into words 
    for j in word_tokenize(i): 
        temp.append(j.lower()) 
  
    data.append(temp) 
  
# Create CBOW model 
model1 = gensim.models.Word2Vec(data, min_count = 1,  
                              size = 100, window = 5) 
  
x1 = model1.similarity('orange', 'eat')
x2 = model1.similarity('alice', 'eat')
x3 = model1.similarity('orange', 'then')
x4 = model1.similarity('orange', 'no')
x5 = model1.similarity('woman', 'sir')
x6 = model1.similarity('can', 'eat')
x7 = model1.similarity('project', 'eat')
x8 = model1.similarity('alice', 'bad')
x9 = model1.similarity('no', 'why')
x10 = model1.similarity('alice', 'wonderful')
      
```

```
te=[x1,x2,x3,x4,x5,x6,x7,x8,x9,x10]
```

```
model2 = gensim.models.Word2Vec(data, min_count = 1, size = 100, 
                                             window = 5, sg = 1) 
  
# Print results 
x11 = model2.similarity('orange', 'eat')
x12 = model2.similarity('alice', 'eat')
x13 = model2.similarity('orange', 'then')
x14 = model2.similarity('orange', 'no')
x15 = model2.similarity('woman', 'sir')
x16 = model2.similarity('can', 'eat')
x17 = model2.similarity('project', 'eat')
x18 = model2.similarity('alice', 'bad')
x19 = model2.similarity('no', 'why')
x20 = model2.similarity('alice', 'wonderful')
```

```
globa= [x11,x12,x13,x14,x15,x16,x17,x18,x19,x20]
```

```
resulting_file=list(zip(globa,te))
```
